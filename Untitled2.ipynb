In this project, we explore how effective modern regression techniques are at modeling stock price behavior â€” specifically Tesla's (TSLA) closing price.

Instead of using linear regression (which is far too simplistic for financial data), we compare:

- **Support Vector Regression (SVR)** â€” good for capturing nonlinear trends
- **XGBoost** â€” a powerful gradient boosting model that handles noise and complex patterns well

We'll evaluate both models using:
- Mean Squared Error (MSE)
- RÂ² Score
- Visual comparison with actual stock prices

## ðŸ“Š Step 1: Collect Historical Stock Price Data

We'll fetch Tesla (TSLA) closing prices for the past 90 days using the `yfinance` library. This gives us real-world data to train our models on.

```python
import yfinance as yf
from datetime import datetime, timedelta
import pandas as pd
import numpy as np

# Define date range
end_date = datetime.today()
start_date = end_date - timedelta(days=90)

# Download data
tsla = yf.download('TSLA', start=start_date, end=end_date)
tsla = tsla[['Close']].dropna().reset_index()
tsla['Day'] = np.arange(len(tsla))  # Add time index
```

## ðŸ§¹ Step 2: Prepare Data for Regression

We'll use:
- `Day` index as the input (X)
- `Close` price as the target (y)

SVR and XGBoost both benefit from scaled input features, so we'll normalize the day index using `StandardScaler`.


```python
from sklearn.preprocessing import StandardScaler

X = tsla[['Day']].values
y = tsla['Close'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## ðŸ¤– Step 3: Train SVR and XGBoost Regressors

Weâ€™ll now train:
- **SVR** with RBF kernel â€” ideal for modeling smooth nonlinear changes
- **XGBoost** â€” known for outperforming most traditional regressors on tabular data

Both will be trained on scaled input.

```python
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = {
    "SVR (RBF kernel)": SVR(kernel='rbf', C=100, epsilon=1.0),
    "XGBoost": XGBRegressor(n_estimators=100, learning_rate=0.1)
}

results = {}

for name, model in models.items():
    model.fit(X_scaled, y)
    preds = model.predict(X_scaled)

    mse = mean_squared_error(y, preds)
    r2 = r2_score(y, preds)

    results[name] = {'preds': preds, 'MSE': mse, 'R2': r2}

    print(f"âœ… {name}")
    print(f"   MSE: {mse:.4f}")
    print(f"   RÂ²:  {r2:.4f}\n")
```

## ðŸ“ˆ Step 4: Visualize Actual vs Predicted Prices

Letâ€™s compare the predictions from SVR and XGBoost to the actual TSLA stock prices to visually assess performance.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(X, y, label='Actual Price', color='black', linewidth=2)

for name, res in results.items():
    plt.plot(X, res['preds'], label=name)

plt.title("Tesla Stock Price Prediction")
plt.xlabel("Day Index")
plt.ylabel("Closing Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```

## âœ… Step 5: Conclusion

### ðŸ” Summary:
- **SVR** captures general trend well, though struggles slightly with sharp turns.
- **XGBoost** models the price curve with better fidelity, making it the stronger choice for this dataset.
